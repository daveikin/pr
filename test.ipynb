```python
import numpy as np
AttScores=np.array([0.1,0.5,0.3, 0.1]) 
Input= np.array([[1,0,2],[0,1,3],[1,3,0],[0,0,0]])
print(AttScores)
print(Input)
print(AttScores@Input)
```

```python
from scipy.special import softmax
Query=np.array([0,0,1])
UnnormScores=Input@Query
AttScores=softmax(UnnormScores)
print(AttScores)
print(Input)
print(AttScores@Input)
```
